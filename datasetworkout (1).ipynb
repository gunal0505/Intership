```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/testt.csv')
```

```python
df['department'] = df['department'].astype(str)
```

```python
print(df['department'].dtypes)
```

```python
sns.countplot()
```

```python
sns.set(style = 'darkgrid', rc = {'figure.figsize':(10, 6)}, font_scale = 1.5)
```

```python
mf=df['age'].std()
print(mf)
```

```python
sns.histplot(df['education'])
```

```python
sns.lineplot(data=df.sort_values('education'),x='education',y='age')
plt.title('Education  Age')
plt.show()
```

```python
sns.jointplot(x = 'education', y = 'department', data = df)
plt.xticks(rotation=45)
plt.title('Education Department')
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='department')
```
